```python
import json
import matplotlib.pyplot as plt

# 학습 데이터 로드
with open('2025_07_08/exp3_learning_progress.json', 'r') as f:
    data = json.load(f)

# Reward 추이
plt.figure(figsize=(10, 6))
plt.plot(data['reward_history'])
plt.title('EXP3 Learning Progress')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.show()

# 상위 Arms
weights = data['weights']
top_10 = sorted(enumerate(weights), key=lambda x: x[1], reverse=True)[:10]
for rank, (arm_idx, weight) in enumerate(top_10):
    print(f"{rank+1}. Arm {arm_idx}: weight={weight:.4f}")
```
